```python
import pandas as pd
data = [[77,56,34,67,41,56,78,12],
        [58,40,24,38,14,66,54,44],
        [74,32,41,76,13,30,40,45],
        [4,16,66,47,1,52,78,12],
        [14,0,74,28,17,58,52,24]]
df = pd.DataFrame(data , columns=["OOAD","Data Science","Statistics","Python","Sci","Maths","c#","Java"] , index = ["Mahek","Diya","Shweta","Vraj","Het"])
df
```

```python
fig , axes = plt.subplots(2,2, layout="constrained")

axes[0,0].plot(df.index,df.Python,'k--' , label='py')
axes[0,0].set_xlabel('X axis')
axes[0,0].set_ylabel('Y axis')
axes[0,0].set_title('Chart')
axes[0,0].legend()

axes[0,1].plot(df.index,df.Sci,'g--' , label='sci')
axes[0,1].set_xlabel('X axis')
axes[0,1].set_ylabel('Y axis')
axes[0,1].set_title('Chart')
axes[0,1].legend()

axes[1,0].plot(df.index,df.OOAD,'c--' , label='ooad')
axes[1,0].set_xlabel('X axis')
axes[1,0].set_ylabel('Y axis')
axes[1,0].set_title('Chart')
axes[1,0].legend()

axes[1,1].plot(df.index,df.Java,'m--' , label='java')
axes[1,1].set_xlabel('X axis')
axes[1,1].set_ylabel('Y axis')
axes[1,1].set_title('Chart')
axes[1,1].legend()


fig.savefig("filename.png")
```
